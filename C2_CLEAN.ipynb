UCS Calculation

```python
shale_ranges = [
    (2550, 2830),
    (2860,2880),
    (2900,2920),
    (2960,2970),
    (2990, 3060),
    (3070, 3100)
]

sand_ranges = [
    (2830, 2960),
    (2880,2900),
    (2920, 2960),
    (2970, 2990),
    (3060, 3070)
]
```

```python
import numpy as np

df_reservoir["UCS"] = np.nan

# ---------------- SHALE ----------------
for start, end in shale_ranges:
    mask = (df_reservoir["DEPTH"] >= start) & (df_reservoir["DEPTH"] < end)

    phi = df_reservoir.loc[mask, "NPHI"]

    df_reservoir.loc[mask, "UCS"] = 1.001 * (phi ** (-1.143))


# ---------------- SANDSTONE ----------------
for start, end in sand_ranges:
    mask = (df_reservoir["DEPTH"] >= start) & (df_reservoir["DEPTH"] < end)

    phi = df_reservoir.loc[mask, "NPHI"]

    # clip porosity
    phi = phi.clip(lower=0.2, upper=0.35)

    df_reservoir.loc[mask, "UCS"] = 254 * (1 - 2.7 * phi) ** 2
```

```python
import numpy as np

# -----------------------------------
# Constants
# -----------------------------------
g = 9.81  # m/s²

# -----------------------------------
# Convert density to kg/m3
# assuming RHO8 is in g/cc
# -----------------------------------
rho = df['RHO8'] * 1000

# -----------------------------------
# Depth increment
# -----------------------------------
df['dZ'] = df['DEPTH'].diff()

# first depth starts from 1168.146 m
df.loc[df.index[0], 'dZ'] = 1168.146

# -----------------------------------
# Vertical stress gradient increment
# dSv = rho * g * dz
# -----------------------------------
dSv = rho * g * df['dZ']

# cumulative sum -> Pa
df['Sv'] = dSv.cumsum()

# convert to MPa
df['Sv'] = df['Sv'] / 1e6

# -----------------------------------
# Hydrostatic pressure
# dPhyd = 1000 * g * dz
# -----------------------------------
dPhyd = 1000 * g * df['dZ']

# cumulative sum -> Pa
df['Phyd'] = dPhyd.cumsum()

# convert to MPa
df['Phyd'] = df['Phyd'] / 1e6

# -----------------------------------
# Check
# -----------------------------------
print(df[['DEPTH', 'Sv', 'Phyd']].head())
```

#Pore pressure using Eatons method

**NCT FOR SHALE ZONE 1**

```python
import numpy as np
import matplotlib.pyplot as plt

# select interval
zone = df_reservoir[
    (df_reservoir["DEPT"] >= 2551) &
    (df_reservoir["DEPT"] <= 2830)
].copy()

depth = zone["DEPT"].values
vp = zone["Vp"].values

# remove invalid values
mask = (~np.isnan(vp)) & (vp > 0)
depth = depth[mask]
vp = vp[mask]

# fit ln(Vp) = ln(a) + b*z
coef = np.polyfit(depth, np.log(vp), 1)

b = coef[0]
a = np.exp(coef[1])

print("a =", a)
print("b =", b)

# predicted NCT
vp_nct = a * np.exp(b * depth)

# save permanently
df_reservoir["Vp_NCT"] = np.nan

mask_main = (
    (df_reservoir["DEPT"] >= 2551) &
    (df_reservoir["DEPT"] <= 2830)
)

df_reservoir.loc[mask_main, "Vp_NCT"] = (
    a * np.exp(b * df_reservoir.loc[mask_main, "DEPT"])
)

# plot
plt.figure(figsize=(6,10))
plt.plot(vp, depth, label="Actual Vp", lw=2)
plt.plot(vp_nct, depth, "--", label="Exponential NCT", lw=2)

plt.gca().invert_yaxis()
plt.xlabel("Vp")
plt.ylabel("Depth (m)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

**Calculating Pp eaton shale zone 1**

```python
import numpy as np

# Eaton exponent for velocity
n = 3

# initialize column
df_reservoir["Pp_Eaton"] = df_reservoir["Phyd"]

# mask: depth interval + overpressure condition
mask = (
    (df_reservoir["DEPT"] >= 2551) &
    (df_reservoir["DEPT"] <= 2830) &
    (df_reservoir["Vp_NCT"] > df_reservoir["Vp"])
)

# Eaton calculation
df_reservoir.loc[mask, "Pp_Eaton"] = (
    df_reservoir.loc[mask, "Sv"]
    - (
        df_reservoir.loc[mask, "Sv"]
        - df_reservoir.loc[mask, "Phyd"]
    )
    * (
        df_reservoir.loc[mask, "Vp"]
        / df_reservoir.loc[mask, "Vp_NCT"]
    )**n
)

print("Calculated points:", mask.sum())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ==========================================
# 1. Create shale zone dataframe
# ==========================================

df_temp = df_reservoir[
    (df_reservoir['DEPTH'] >= 2550) &
    (df_reservoir['DEPTH'] <= 2830)
].copy()

# Keep only required columns
df_temp = df_temp[['DEPTH', 'NPHI']].dropna()

# Sort by depth
df_temp = df_temp.sort_values('DEPTH')

# ==========================================
# 2. Moving average smoothing
# ==========================================

df_temp['NPHI_smooth'] = (
    df_temp['NPHI']
    .rolling(window=25, center=True, min_periods=1)
    .mean()
)

# ==========================================
# 3. Exponential NCT fit
# ==========================================

def porosity_nct(depth, phi0, c):
    return phi0 * np.exp(-c * depth)

# Fit curve
popt, _ = curve_fit(
    porosity_nct,
    df_temp['DEPTH'],
    df_temp['NPHI_smooth'],
    p0=[0.8, 0.0005],
    maxfev=10000
)

phi0_fit, c_fit = popt

# Generate NCT
df_temp['NPHI_NCT'] = porosity_nct(
    df_temp['DEPTH'],
    phi0_fit,
    c_fit
)

print(f"phi0 = {phi0_fit:.4f}")
print(f"c    = {c_fit:.6f}")

# ==========================================
# 4. Plot
# ==========================================

plt.figure(figsize=(7,10))

# Raw NPHI
plt.plot(
    df_temp['NPHI'],
    df_temp['DEPTH'],
    color='lightgray',
    linewidth=1,
    label='Raw NPHI'
)

# Smoothed NPHI
plt.plot(
    df_temp['NPHI_smooth'],
    df_temp['DEPTH'],
    color='blue',
    linewidth=2,
    label='Moving Avg NPHI'
)

# NCT
plt.plot(
    df_temp['NPHI_NCT'],
    df_temp['DEPTH'],
    color='red',
    linewidth=3,
    label='Porosity NCT'
)

plt.gca().invert_yaxis()

plt.xlabel('Porosity (NPHI)')
plt.ylabel('Depth (m)')
plt.title('Porosity NCT : Shale Zone 1 (2550–2830 m)')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np

# ==========================================
# Parameters from porosity NCT fit
# ==========================================

phi0 = 0.48
c = 0.002191

# ==========================================
# Create temporary shale-zone dataframe
# ==========================================

df_temp1 = df_reservoir[
    (df_reservoir['DEPTH'] >= 2550) &
    (df_reservoir['DEPTH'] <= 2830)
].copy()

# Relative depth
df_temp1['Z_rel'] = (
    df_temp1['DEPTH'] - df_temp1['DEPTH'].min()
)

# Avoid division by zero
df_temp1['Z_rel'] = df_temp1['Z_rel'].replace(0, 0.1)

# Initialize with hydrostatic pressure
df_temp1['Pp_porosity'] = df_temp1['Phyd']

# Apply only where actual porosity > NCT
mask = df_temp1['NPHI'] > df_temp1['NPHI_NCT']

# Safe porosity values
phi_safe = np.clip(df_temp1['NPHI'], 1e-6, None)

# ==========================================
# Porosity-based pore pressure
# ==========================================

df_temp1.loc[mask, 'Pp_porosity'] = (
    df_temp1.loc[mask, 'Sv']
    -
    (
        (
            df_temp1.loc[mask, 'Sv']
            -
            df_temp1.loc[mask, 'Phyd']
        )
        *
        (
            (
                np.log(phi0)
                -
                np.log(phi_safe[mask])
            )
            /
            (
                c * df_temp1.loc[mask, 'Z_rel']
            )
        )
    )
)

# Lower bound = hydrostatic pressure
df_temp1['Pp_porosity'] = np.maximum(
    df_temp1['Pp_porosity'],
    df_temp1['Phyd']
)

# ==========================================
# Save back into df_reservoir
# ==========================================

# Create empty column if not present
df_reservoir['Pp_porosity'] = np.nan

# Save calculated values back
df_reservoir.loc[
    df_temp1.index,
    'Pp_porosity'
] = df_temp1['Pp_porosity']

print("Pp_porosity saved into df_reservoir")
```

**Calculating Pp Shale zone 2**

```python
import numpy as np

# ==========================================
# Parameters
# ==========================================

phi0 = 0.38240
c = 0.0021

# ==========================================
# Depth interval
# ==========================================

mask_interval = (
    (df_reservoir['DEPTH'] >= 3020) &
    (df_reservoir['DEPTH'] <= 3098)
)

# Relative depth
z_rel = (
    df_reservoir.loc[mask_interval, 'DEPTH']
    -
    3020
)

# Avoid division by zero
z_rel = z_rel.replace(0, 0.1)

# ==========================================
# Initialize with hydrostatic pressure
# ==========================================




# ==========================================
# Apply only where actual porosity > NCT
# ==========================================

mask_overpressure = (
    mask_interval
    &
    (
        df_reservoir['NPHI']
        >
        df_reservoir['NPHI_NCT']
    )
)

# Safe porosity
phi_safe = np.clip(
    df_reservoir.loc[mask_overpressure, 'NPHI'],
    1e-6,
    None
)

# ==========================================
# Calculate pore pressure
# ==========================================

df_reservoir.loc[mask_overpressure, 'Pp_porosity'] = (
    df_reservoir.loc[mask_overpressure, 'Sv']
    -
    (
        (
            df_reservoir.loc[mask_overpressure, 'Sv']
            -
            df_reservoir.loc[mask_overpressure, 'Phyd']
        )
        *
        (
            (
                np.log(phi0)
                -
                np.log(phi_safe)
            )
            /
            (
                c * z_rel.loc[mask_overpressure]
            )
        )
    )
)

# Lower bound = hydrostatic pressure
df_reservoir.loc[mask_interval, 'Pp_porosity'] = np.maximum(
    df_reservoir.loc[mask_interval, 'Pp_porosity'],
    df_reservoir.loc[mask_interval, 'Phyd']
)

print("Pp_porosity calculated and saved in df_reservoir")
```
