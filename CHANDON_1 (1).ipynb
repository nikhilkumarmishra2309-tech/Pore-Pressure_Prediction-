```python
import lasio
import pandas as pd

# Load LAS file
las = lasio.read('/content/BestDT_final.las')

# Convert to dataframe
df_1 = las.df().reset_index()

# View first rows
print(df_1.head())
```

```python
df_1=df_1[['DTSM','DEPT','DTCO']]
```

```python
df_1=df_1.dropna()
```

```python
df_1.isnull().sum()
```

```python
df_1['Vp']=1e6/df_1['DTCO']
df_1['Vs']=1e6/df_1['DTSM']
```

```python
df_1.describe()
```

```python
df_1.columns
```

```python
import matplotlib.pyplot as plt

# Plot Vp and Vs with Depth
plt.figure(figsize=(8, 12))

plt.plot(df_1['Vp'], df_1['DEPT'], label='Vp')
plt.plot(df_1['Vs'], df_1['DEPT'], label='Vs')

# Depth increasing downward
plt.gca().invert_yaxis()

plt.xlabel('Velocity (m/s)')
plt.ylabel('Depth')
plt.title('Vp and Vs vs Depth')
plt.legend()
plt.grid(True)

plt.show()
```

```python
df_1=df_1[['DEPT','Vp']]
```

```python
# Apply Vp clipping based on depth intervals

# 2200 - 2400 -> 2800
mask1 = (df_1['DEPT'] >= 2200) & (df_1['DEPT'] < 2400)
df_1.loc[mask1, 'Vp'] = df_1.loc[mask1, 'Vp'].clip(upper=2800)

# 2400 - 2600 -> 2500
mask2 = (df_1['DEPT'] >= 2400) & (df_1['DEPT'] < 2600)
df_1.loc[mask2, 'Vp'] = df_1.loc[mask2, 'Vp'].clip(upper=2500)

# 2600 - 2800 -> 3800
mask3 = (df_1['DEPT'] >= 2600) & (df_1['DEPT'] < 2800)
df_1.loc[mask3, 'Vp'] = df_1.loc[mask3, 'Vp'].clip(upper=3800)

# 2800 - 3000 -> 3300
mask4 = (df_1['DEPT'] >= 2800) & (df_1['DEPT'] < 3000)
df_1.loc[mask4, 'Vp'] = df_1.loc[mask4, 'Vp'].clip(upper=3300)

print("Vp clipping applied successfully.")
```

```python
import matplotlib.pyplot as plt

# Plot clipped Vp with depth
plt.figure(figsize=(8, 12))

plt.plot(df_1['Vp'], df_1['DEPT'], label='Clipped Vp')

# Depth increasing downward
plt.gca().invert_yaxis()

plt.xlabel('Vp (m/s)')
plt.ylabel('Depth')
plt.title('Clipped Vp vs Depth')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Keep only depth interval 2102 m to 3017 m
df_1 = df_1[(df_1['DEPT'] >= 2102) & (df_1['DEPT'] <= 3017)].copy()

# Apply moving average filter with window = 15 on Vp
df_1['Vp_smooth'] = df_1['Vp'].rolling(window=15, center=True).mean()

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 12))

plt.plot(df_1['Vp'], df_1['DEPT'], label='Original Vp', alpha=0.5)
plt.plot(df_1['Vp_smooth'], df_1['DEPT'], label='Smoothed Vp', linewidth=2)

# Depth increasing downward
plt.gca().invert_yaxis()

plt.xlabel('Vp (m/s)')
plt.ylabel('Depth')
plt.title('Vp Moving Average Filter (Window = 15)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot GR vs Depth
plt.figure(figsize=(6,10))

plt.plot(df_2['GR'], df_2['DEPT'])

# In well logs, depth increases downward
plt.gca().invert_yaxis()

plt.xlabel('GR')
plt.ylabel('Depth')
plt.title('GR vs Depth')

plt.grid(True)
plt.show()
```

```python
# Apply clipping only between depth 2800 and 3000

mask = (df_2['DEPT'] >= 2800) & (df_2['DEPT'] <= 3000)

# Clip GR values to maximum 150 in that interval
df_2.loc[mask, 'GR'] = df_2.loc[mask, 'GR'].clip(upper=150)

# Plot after clipping
import matplotlib.pyplot as plt

plt.figure(figsize=(6,10))

plt.plot(df_2['GR'], df_2['DEPT'])

plt.gca().invert_yaxis()

plt.xlabel('GR')
plt.ylabel('Depth')
plt.title('GR vs Depth After Clipping')

plt.grid(True)
plt.show()
```

```python
# Apply moving average filter with window size 15 on GR

df_2['GR_smooth'] = (
    df_2['GR']
    .rolling(window=15, center=True)
    .mean()
)

# Plot original and smoothed GR

import matplotlib.pyplot as plt

plt.figure(figsize=(6,10))

plt.plot(df_2['GR'], df_2['DEPT'], label='Original GR', alpha=0.5)
plt.plot(df_2['GR_smooth'], df_2['DEPT'], label='Smoothed GR', linewidth=2)

plt.gca().invert_yaxis()

plt.xlabel('GR')
plt.ylabel('Depth')
plt.title('GR Moving Average Filter (Window = 15)')

plt.legend()
plt.grid(True)

plt.show()
```

```python
# Apply moving average only from depth 2700 onward

mask = df_2['DEPT'] >= 2700

# Keep original GR first
df_2['GR_smooth_30'] = df_2['GR']

# Apply rolling mean only on selected interval
df_2.loc[mask, 'GR_smooth_30'] = (
    df_2.loc[mask, 'GR']
    .rolling(window=60, center=True)
    .mean()
)

# Plot

import matplotlib.pyplot as plt

plt.figure(figsize=(6,10))

plt.plot(df_2['GR'], df_2['DEPT'], label='Original GR', alpha=0.5)
plt.plot(df_2['GR_smooth_30'], df_2['DEPT'], label='Window 30 Smooth', linewidth=2)

plt.gca().invert_yaxis()

plt.xlabel('GR')
plt.ylabel('Depth')
plt.title('Selective Moving Average Smoothing')

plt.legend()
plt.grid(True)

plt.show()
```

```python
# Apply clipping on RHO8 for different depth intervals
# Lower limit everywhere = 1.90

# 2200 - 2400  --> upper = 2.25
mask1 = (df_2['DEPT'] >= 2200) & (df_2['DEPT'] < 2400)
df_2.loc[mask1, 'RHO8'] = df_2.loc[mask1, 'RHO8'].clip( upper=2.25)

# 2400 - 2600 --> upper = 2.25
mask2 = (df_2['DEPT'] >= 2400) & (df_2['DEPT'] < 2600)
df_2.loc[mask2, 'RHO8'] = df_2.loc[mask2, 'RHO8'].clip( upper=2.25)

# 2600 - 2800 --> upper = 2.75
mask3 = (df_2['DEPT'] >= 2600) & (df_2['DEPT'] < 2800)
df_2.loc[mask3, 'RHO8'] = df_2.loc[mask3, 'RHO8'].clip( upper=2.75)

# 2800 - 3000 --> upper = 2.45
mask4 = (df_2['DEPT'] >= 2800) & (df_2['DEPT'] <= 3000)
df_2.loc[mask4, 'RHO8'] = df_2.loc[mask4, 'RHO8'].clip(lower=1.90, upper=2.45)
# 3000 - 3000 --> upper = 2.45
mask4 = (df_2['DEPT'] >= 3000)
df_2.loc[mask4, 'RHO8'] = df_2.loc[mask4, 'RHO8'].clip(lower=2.0, upper=2.50)


# Plot RHO8 vs Depth

import matplotlib.pyplot as plt

plt.figure(figsize=(6,10))

plt.plot(df_2['RHO8'], df_2['DEPT'])

plt.gca().invert_yaxis()

plt.xlabel('RHO8')
plt.ylabel('Depth')
plt.title('RHO8 After Clipping')

plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# use combined vp log
data = df_vp[['DEPTH', 'Vp']].dropna()

x = data['DEPTH'].values
y = data['Vp'].values

# log transform
log_y = np.log(y)

# linear fit: ln(Vp) = ln(a) + b * depth
b, ln_a, r, p, err = linregress(x, log_y)

a = np.exp(ln_a)

print("NCT Fit Results:")
print("a =", a)
print("b =", b)
```

```python
# extend depth range to include prediction at 1300 m
depth_extended = np.linspace(1300, 3019, 500)

vp_pred = a * np.exp(b * depth_extended)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,10))

# original data
plt.plot(df_vp['Vp'], df_vp['DEPTH'], '.', label='Observed Vp')

# fitted trend
plt.plot(vp_pred, depth_extended, 'r', label='NCT Fit')

plt.gca().invert_yaxis()
plt.xlabel('Vp (m/s)')
plt.ylabel('Depth (m)')
plt.legend()
plt.title('Vp = a e^(b*Depth) Trend Fit')
plt.grid()

plt.show()
```

# Estimating Sv and Phyd from 1300m to 3018m

```python
import numpy as np

df_rhob['Phyd'] = 1000 * 9.8 * df_rhob['DEPTH']
```

```python
df_rhob['RHO8']=df_rhob['RHO8']*1000
```

```python
sv = []
sv_val = 0

depth = df_rhob['DEPTH'].values
rho = df_rhob['RHO8'].values

for i in range(len(df_rhob)):
    if depth[i] <= 1300:
        sv_val = 1000 * 9.8 * depth[i]
    else:
        if i == 0:
            continue
        dh = depth[i] - depth[i-1]
        sv_val = sv_val + rho[i] * 9.8 * dh

    sv.append(sv_val)

df_rhob['SV'] = sv
```
